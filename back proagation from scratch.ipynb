```python
import numpy as np
from numpy.random import rand
```

```python
x=np.array(([9,8],[6,3],[9,1],[9,8])) #features
```

```python
y=np.array([[0],[1],[1],[0]]) #labels
```

```python
def sigmoid(z) :
    return 1/(1+np.exp(-z))
    
```

```python
class Neaurnetwork:
    def __init__(self,x,y,nodes_in_layer1=4,nodes_in_layer2=3,nodes_in_layer3=1,l_rate=1):
        #define x,y
        self.input_in_layer0 = x
        self.y=y
        self.l_rate=l_rate
        
        #number of neurons in layer
        self.nodes_in_layer1=nodes_in_layer1
        self.nodes_in_layer2=nodes_in_layer2
        self.nodes_in_layer3=nodes_in_layer3
        
        #set the weights (theta) metrics
        self.theta_layer0=np.random.rand(self.input_in_layer0.shape[1]+1,self.nodes_in_layer1)
        self.theta_layer1=np.random.rand(self.nodes_in_layer1+1,self.nodes_in_layer2)
        self.theta_layer2=np.random.rand(self.nodes_in_layer2+1,self.nodes_in_layer3)
  

        
    def feedforward(self):
        #compute all nodes (a1,a2,a3,a4) in layer 1
        self.z1=self.theta_layer0[0]+np.dot(self.input_in_layer0,self.theta_layer0[1:])
        self.layer1=sigmoid(self.z1)
        
         #compute all nodes (a1,a2,a3) in layer 2
        self.z2=self.theta_layer1[0]+np.dot(self.layer1,self.theta_layer1[1:])
        self.layer2=sigmoid(self.z2)
        
         #compute all nodes (a1,a2,a3) in layer 1
        self.z3=self.theta_layer2[0]+np.dot(self.layer2,self.theta_layer2[1:])
        self.layer3=sigmoid(self.z3)
        
        return self.layer3
    
    ### cost function
    
    def cost_fun(self):
        self.n=self.input_in_layer0.shape[0] #number of training example
        self.cost=(1/self.n)*np.sum(-self.y *np.log(self.layer3)-(1-self.y)*np.log(1-self.layer3))#cross entropy
        return self.cost
    
    ### back propgation
    def backpropgation(self):
        
        ## dervation of E with respect to theta and bias in layer2
        self.dE_layer3 =(1/self.n)*(self.layer3-y)/(self.layer3*(1-self.layer3))
        self.dE_dz3 =np.multiply(self.dE_layer3,sigmoid(self.z3)*(1-sigmoid(self.z3)))
        self.dE_dtheta2=np.dot(self.layer2.T, self.dE_dz3)
        self.dbias2=np.dot(np.ones(self.n), self.dE_dz3)
        
        ## dervation of E with respect to theta and bias in layer1
        self.dE_layer2 =np.dot(self.dE_dz3,self.theta_layer2[1:].T)
        self.dE_dz2 =np.multiply(self.dE_layer2,sigmoid(self.z2)*(1-sigmoid(self.z2)))
        self.dE_dtheta1=np.dot(self.layer1.T, self.dE_dz2)
        self.dbias1=np.dot(np.ones(self.n), self.dE_dz2)
        
        
        ## dervation of E with respect to theta and bias in layer0
        self.dE_layer1 =np.dot(self.dE_dz2,self.theta_layer1[1:].T)
        self.dE_dz1 =np.multiply(self.dE_layer1,sigmoid(self.z1)*(1-sigmoid(self.z1)))
        self.dE_dtheta0=np.dot(self.input_in_layer0.T, self.dE_dz1)
        self.dbias0=np.dot(np.ones(self.n), self.dE_dz1)
        
        ## updating theta using gradient descent in layer 2,1 and 0
        self.theta_layer2[1:] = self.theta_layer2[1:]- self.l_rate * self.dE_dtheta2
        self.theta_layer1[1:] = self.theta_layer1[1:]- self.l_rate * self.dE_dtheta1
        self.theta_layer0[1:] = self.theta_layer0[1:]- self.l_rate * self.dE_dtheta0
        
        
        
        ## updating bias using gradient descent in layer 2,1 and 0
        self.theta_layer2[0] = self.theta_layer2[0] - self.l_rate * np.sum(self.dE_dz3, axis=0)
        self.theta_layer1[0] = self.theta_layer1[0] - self.l_rate * np.sum(self.dE_dz2, axis=0)
        self.theta_layer0[0] = self.theta_layer0[0] - self.l_rate * np.sum(self.dE_dz1, axis=0)
        
        
        return self
        
        
        
```

```python

NN=Neaurnetwork(x,y)
```

### feed forward

```python
predict=NN.feedforward()
```
